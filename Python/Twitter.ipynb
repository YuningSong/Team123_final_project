### Scraper of hashtags

```python
most_dict = {
'Name':[],
'userid':[],
'mostPopHashtag1' : [],
'mostPopHashtag2' : [],
'mostPopHashtag3' : [],
'mostPopHashtag4' : [],
'mostPopHashtag5' : [],
'mostPopPost':[]
}
for x in name1[:100]:
    user = api.get_user(x)
    end_date = datetime.utcnow() - timedelta(days=90)
    status = []
    for x in Cursor(api.user_timeline, id=user.id).items():
        if x.created_at > end_date:
            status.append(x)
        if x.created_at <= end_date:
            break
    hashtagInter=[]
    for j in status:
        internal = []
        internal =  j.entities.get("hashtags")
        for h in internal:
            if(len(h)==0):
                hashtagInter.append("null")
            else: 
                hashtagInter.append(h['text'])

    for i in status:
        hashtags_dict={}
        hashtags = i.entities.get('hashtags')
        for hashtag in hashtagInter:
            if hashtag in hashtags_dict.keys():
                hashtags_dict[hashtag] += 1
            else:
                hashtags_dict[hashtag] = 1
    if len(sorted(hashtags_dict, key=hashtags_dict.get, reverse=True))<5:
        continue
    else :
        most_dict['Name'].append(x) 
        most_dict['userid'].append(user.id)        
        most_dict['mostPopHashtag1'].append(sorted(hashtags_dict, key=hashtags_dict.get, reverse=True)[0])
        most_dict['mostPopHashtag2'].append(sorted(hashtags_dict, key=hashtags_dict.get, reverse=True)[1])
        most_dict['mostPopHashtag3'].append(sorted(hashtags_dict, key=hashtags_dict.get, reverse=True)[2])
        most_dict['mostPopHashtag4'].append(sorted(hashtags_dict, key=hashtags_dict.get, reverse=True)[3])
        most_dict['mostPopHashtag5'].append(sorted(hashtags_dict, key=hashtags_dict.get, reverse=True)[4])

    post_favorites = {}
    for i in status:
        post_favorites[i.text] = i.favorite_count

    most_dict['mostPopPost'].append(sorted(post_favorites, key=post_favorites.get, reverse=True)[:1])
```
